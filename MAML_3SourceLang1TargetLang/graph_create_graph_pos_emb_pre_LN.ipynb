```python
import matplotlib.pyplot as plt

f1 = open( './train3.log')
f2 = open( './val3.log')

i = 0
x = []
loss = []
error = []
bleu = []
v_i = 0
v_x = []
v_loss = []
v_error = []
v_bleu = []

for line1 in f1:
    #print( line1 )
    line1_split = line1.split(", ")
    #print( line1_split[3])
    i += 1
    x.append( i )
    loss.append( float(line1_split[2].split(' ')[1]) )
    error.append( float(line1_split[3].split(' ')[1]) )
    #bleu.append( float(line1_split[3].split( ' ' )[1]))
    #line1 = f1.readline()
    #line1 = line1.replace( "\n", "")
    
for line2 in f2:
    #print( line1 )
    line2_split = line2.split(", ")
    #print( line2_split)
    v_i += 1
    v_x.append( (v_i ) * 1000 )
    v_loss.append( float(line2_split[2].split(' ')[1]) )
    #print( line2_split[2].split(' ')[2])
    v_error.append( float(line2_split[3].split(' ')[1] ) )
    #v_bleu.append( float(line2_split[3].split( ' ' )[1]))
    #line2 = f2.readline()
    #line2 = line2.replace( "\n", "")

#print( v_error )
    
plt.plot( x, loss, label="Train loss")
plt.plot( v_x, v_loss, label="Val Loss" )
plt.title( "Loss")
plt.xlabel( 'steps')
plt.ylabel( 'Loss')
plt.legend()
plt.ylim( 1, 5 )
#plt.xlim( 194000, 200000)
#plt.xlim( 198000, 200500)
plt.show()

plt.plot( x, error, label = 'Train WER')
plt.plot( v_x, v_error, label="Val WER")
plt.title( "WER")
plt.xlabel( 'steps')
plt.ylabel( 'WER')
plt.legend()
plt.ylim( 0.3, 1 )
#plt.xlim( 104000, 120000)
#plt.xlim( 0, 210000)
#plt.xlim( 194000, 210000)
plt.show()

#plt.plot( x, bleu, label = 'Train BLEU')
#plt.plot( v_x, v_bleu, label="Val BLEU")
#plt.title( "BLEU")
#plt.xlabel( 'Epochs')
#plt.ylabel( 'BLEU')
#plt.legend()
#plt.ylim( 0, 20)
#plt.show()


#plt.plot( v_x, v_loss)
#plt.title( "Validation Loss")
#plt.xlabel( 'Epochs')
#plt.ylabel( 'Loss')

#plt.show()
```
